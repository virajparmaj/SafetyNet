##Moving all files into main folders

```python
import os
import shutil

# Path to the main folder
main_folder = '/content/Dataset/Updated/Be_Safe'

# List all subfolders in the main folder
subfolders = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

# Iterate over the subfolders and move their contents to the main folder
for subfolder in subfolders:
    subfolder_path = os.path.join(main_folder, subfolder)

    # Move each file in the subfolder to the main folder
    for item in os.listdir(subfolder_path):
        item_path = os.path.join(subfolder_path, item)
        shutil.move(item_path, main_folder)

    # Remove the now empty subfolder
    os.rmdir(subfolder_path)

print("Contents moved successfully!")
```

##CONVERT IMAGE FRAMES INTO A VIDEO

```python
import cv2
import os

main_folder = '/content/Dataset/content/Dataset/Updated/Be_Safe'
out_path = '/content/Dataset/Converted/Watch_Out'

subfolders = os.listdir(main_folder)

for subfolder_name in subfolders:
    subfolder_path = os.path.join(main_folder, subfolder_name)

    if not os.path.isdir(subfolder_path):
        continue

    out_video_name = subfolder_name + '.mp4'
    out_video_full_path = os.path.join(out_path, out_video_name)

    pre_imgs = os.listdir(subfolder_path)
    img = []

    for i in pre_imgs:
        i = os.path.join(subfolder_path, i)
        img.append(i)

    cv2_fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    frame = cv2.imread(img[0])
    size = list(frame.shape)
    del size[2]
    size.reverse()

    # Calculate desired video duration
    desired_duration = 3.0  # in seconds
    total_frames = len(img)
    fps = total_frames / desired_duration

    video = cv2.VideoWriter(out_video_full_path, cv2_fourcc, fps, size)

    for i in range(len(img)):
        video.write(cv2.imread(img[i]))
        print('Frame', i + 1, 'of', len(img), 'in', subfolder_name)

    video.release()
    print('Outputted video to', out_video_full_path)
```
